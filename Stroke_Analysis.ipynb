## The next function uses Seaborn, the package installed earlier, to create a scatterplot illustrating age on the X-axis and average blood glucos level on the Y-Axis.


```python
# Below, I will create a histogram of the age of the participants and their average glucose levels
sns.scatterplot(
    data=df,
    x='age',
    y='avg_glucose_level',
).set(
    xlabel='Age',
    ylabel='Average Glucose Level',
    title='Age vs Average Glucose Level',
);




```

# Analysis:


#### The World Health Organization states that the average blood glucose level for adults when fasting should be within a range of 70mg/dL and 125mg/dL. We can see that while the majority of those studied had levels within that range, as age increasees, those with high blood glucose levels also increases, implying a positive correlation between age and high blood glucose levels. It should be noted that it is not clear the degree to which studied participants fasted.




### Next, I will code a function to create a histogram illustrating age on the X-Axis and Stroke on the Y-Axis, with a clear vertical divide between "Yes" and "No" Stroke responses.



```python
sns.histplot(
    data=df,
    x='age',
    y='stroke'
).set(
    xlabel='Age',
    ylabel='Stroke',
    title='Age vs Stroke',
);



```

# Analysis:


### The darker a blue rectangle is, the more times the corresponding X-Axis and Y-Axis values occur. We see that for those who had a stroke, the darkest blue falls above the range of ages around 80 years old, implying a strong positive correlation between that age range and likelihood of having a stroke.



### Further, we can see that for the rest of the age ranges that reported strokes, the shade of blue is nearly identical. So, we know that the frequency of strokes in all other age ranges that are not 80 that did report having a stroke is nearly identical, implying that no age below 80 had a significantly high report of strokes and therefore no age below 80 is significantly more at risk for having a stroke.


### We must also notice that there are far more (light) blue bars for "Yes" (had a stroke) on the right half of the histogram. The presence of a stroke was more common in ages above 40. Even though the darkest blue bar falls at 80, the presence of a stroke is still prevelant and consistent across ages 40 and above, leading us to draw the conclusion that while ages 80 are most at risk/are at high risk for having a stroke, ages 40 and above are still to be logically considered at risk as well, just to a lesser degree than ages 80 and above.


## The next function will create a histogram illustrating a running total count on the Y-Axis and Urban and Rural residence type on the X-Axis. The blue regions of each bar represent those of the total residence type population who did have a stroke, and the orange regions of each bar represent those of each total residence type population who did not have a stroke.





```python
# Create a histogram of residence only for those who had a stroke 
sns.histplot(
    data=df,
    x='Residence_type',
    hue='stroke',
    multiple='stack',
).set(
    xlabel='Residence Type',
    ylabel='Stroke Report Counts',
    title='Residence Type vs Stroke',
);





```

# Analysis:


### We see that the two bars are almost identical in size and in color distribution. First and foremost, this shows us that their was a near even split between individuals living in both Urban and Rural areas. The color ratio indicates that there are nearly as many people who did have a stroke in Urban areas as there were in Rural areas, and likewise that there are nearly as many people who did not have a stroke in Urban areas as there were in Rural areas. This implies that residence type is not a causing factor for stroke and there is little to no correlation between residence type and likelihood of having a stroke.
